```python
#Task 2
m=input()
q=m[1:len(m)-1]
n=[]
l=""
for i in q :
  if i!="," :
    l+=i
  else :
    n.append(int(l))
    l=""
n+=[int(l)]
if len(n)>=4 :
   ramisa=n[2:len(n)-2:1]
   print(ramisa)
else :
  print("Not Possible")
```

```python
#Task 2 another way
m=input()
l=[]
n=m[1:len(m)-1:1]
g=n.split(",")
for i in g :
  l.append(int(i))
if len(l)>=4 :
   ramisa=l[2:len(l)-2:1]
   print(ramisa)
else :
  print("Not Possible")
```

```python
#Task 6
m=input()
n=[]
l=m.split(",")
for i in l :
  n.append(int(i))
high=0
count=-1
for b in n :
  if b > high :
     high=b
     count+=1
print("My list :" ,n)
print("Largest number in the list is ",high,"which was found at index",count)
```

```python
#Task 6 another way
m=input()
b=[]
n=m.split(",")
for i in n :
   b+=[int(i)]
max=0
index=0
for j in range(len(b)) :
  if b[j]>max :
    max=b[j]
    index=j
print("My list :" , b)
print("Largest number in the list is",max ,"which was found at index",index)
```

```python
#Task 10
m=input()
n=[]
l=m.split(",")
for i in l :
  n.append(int(i))
t=[]
for j in n :
  if j not in t :
    t.append(j)
print("Input list:",n)
print("Modified list :",t)
```

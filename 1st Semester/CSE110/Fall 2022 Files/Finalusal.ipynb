```python
#Task 4 F
def palindrome_checker(m) :
  w=""
  q= ""
  for j in m :
    if j!= " " :
     w+=j
  for i in range(len(w)-1,-1,-1) :
    q+=w[i]
  if w==q :
    return "Palindrome"
  return " Not a palindrome"
print(palindrome_checker(m=input()))

```

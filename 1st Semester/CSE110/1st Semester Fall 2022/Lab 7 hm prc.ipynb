```python
#Task 2
def fibonacci(x) :
    sum = 0
    m=0
    n=1
    while sum <= x :
     print(sum,end=" ")
     m=n
     n=sum
     sum=m+n
fibonacci(10)
```

```python
#Task 7
def show_palindrome(m) :
    c=""
    d=""
    for i in range (1,m+1) :
      c=i
      print(c,end="")
    for b in range(m-1,0,-1) :
      d=b
      print(d,end="")


show_palindrome(m=int(input()))
```

```python
#Task 8
def show_palindrome_triangle(r) :
  space=r-1
  for i in range(1,r+1) :
    for j in range(space) :
      print(" " ,  end="")
    space=space-1
    show_palindrome(i)
    print()
show_palindrome_triangle(5)
```

```python
#Task 10
def make_square(i) :
    d={}
    a,b=i
    for l in range(a,b+1) :
      d[l]=l**2
    return d

print(make_square((1,3)))
```

```python
#Task 11
def rem_duplicate(m) :
    t=[]
    n=list(m)
    for i in n :
      if i not in t :
        t.append(i)
      else :
        continue
    print(t)  
  
rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0))
```

```python
#Task 13
def function_name(m,n,l) :
  if m == "+" :
    q=float(n+l)
    return q
  elif m=="-" :
    w=float(n-l)
    return w
  elif m== "*" :
    t=float(n*l)
    return t
  elif m=="/" :
    i=float(n/l)
    return i
print(function_name(m=input(),n=int(input()),l=int(input())))
```

```python

#Task 14 :
def function_name(s,n) :
    l=""
    q = s[0]
    for i in range(1, len(s)) :
      if i%n!=0 :
        q+=s[i]
      else :
        l+=s[i]
    return q+l

print(function_name(s=input(),n=int(input())))
```

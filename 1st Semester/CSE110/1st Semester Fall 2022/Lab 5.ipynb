```python
#Task6
m=input()
b=[]
n=m.split(",")
for i in n :
   b+=[int(i)]
max=0
index=0
for j in range(len(b)) :
  if b[j]>max :
    max=b[j]
    index=j
print("My list :" , b)
print("Largest number in the list is",max ,"which was found at index",index)
```

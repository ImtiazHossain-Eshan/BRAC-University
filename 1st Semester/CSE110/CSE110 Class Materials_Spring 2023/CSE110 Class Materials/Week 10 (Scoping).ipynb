# **Function Variable Scoping**

```python
# local variable

def func():
    v = 'local variable'
    print(v)
    
func()
print(v) # NameError: name 'v' is not defined
```

```python
# global variable 

v = 'global variable'
def func():
    v = 'local variable' 
    print(v)
    
func()
print(v) 
```

```python
# changing a local variable (change inside the function)

v = 'global variable'

def func():
    v = 'local variable'
    v += ' named v'
    print(v)
    
func()
print(v)
```

```python
# trying to change a global variable

v = 'global variable'
def func():
    v += ' named v' # local variable 'v' referenced before assignment
    print(v)
func()
print(v)

# how to get rid of this situation?
```

```python
# changing a global variable ('global' keyword) 

v = 'global variable'
def func():
    global v
    v += ' named v' # no error this time
    print(v)

func()
print(v)
```

```python
# nested function

v = 'global variable'
def func_out():
    v = 'outer function variable'

    def func_in():
        v = 'inner function variable'
        print(v)
        
    func_in()
    print(v)

func_out()
print(v)
```

```python
# no inner function variable in nested function

v = 'global variable'
def func_out():
    v = 'outer function variable'

    def func_in():
        # v = 'inner function variable'
        print(v)
    func_in()
    print(v)
    
func_out()
print(v)
```

```python
# no outer and inner function variable in nested function

v = 'global variable'
def func_out():
    # v = 'outer function variable'

    def func_in():
        # v = 'inner function variable'
        print(v)
    func_in()
    print(v)
    
func_out()
print(v)
```

```python
# trying to change an outer function variable (inside inner function) in nested function

v = 'global variable'
def func_out():
    v = 'outer function variable'

    def func_in():
        v += ' named v' # local variable 'v' referenced before assignment
        print(v)
    func_in()
    print(v)
    
func_out()
print(v)

# how to get rid of this situation?
```

```python
# changing an outer function variable in nested function ('nonlocal' keyword)

v = 'global variable'
def func_out():
    v = 'outer function variable'

    def func_in():
        nonlocal v
        v += ' named v' # no error this time
        print(v)
    func_in()
    print(v)
    
func_out()
print(v)
```

```python
# changing global variable in nested function

v = 'global variable'
def func_out():
    
    def func_in():
        global v
        v += ' named v' 
        print(v)
        
    func_in()
    print(v)
    
func_out()
print(v) 
```

## Logical Operator

1.   **and** (returns True if both the values or conditions are true, otherwise returns False)

*   False and False = False
*   False and True = False
*   True and False = False
*   True and True = True


2.   **or** (returns False if both the values or conditions are false, otherwise returns True)

*   False or False = False
*   False or True = True
*   True or False = True
*   True or True = True

3.   **not** (returns True if the value or condition is false and vice versa)
*   not False = True
*   not True = False

````python
var1 = True
print(not var1)
````

````python
var1 = False
var2 = True
print(var1 or var2) # True
print(var1 and var2) # False
````

## Compound Boolean Expression (Logical and Comparison Operators Combined)

````python
var1 = 2 < 5
var2 = not True

print(var1 and var2) # False
print(var1 or var2) # True
````

````python
print(13 > 7 and 10 == 9) # True and False = False
print(13 > 7 or 10 == 9) # True or False = True

print(not 10 < 7) # not False = True
````

````python
x = 4

print(x > 2 and x < 5)
# last line can also be written as the following format:
print(2 < x < 5)
````

````python
x = 9

print(x >= 3 and x <= 7)
# last line can also be written as the following format:
print(3 <= x <= 7)
````

## if...elif...(else) statement

````python
money = 500 # 250, 50, 5 

# eat pizza
if money >=300:
    print("Pizza khabo.") 
    money-=300

# eat kacchi
elif money >=200: 
    print("Kacchi khabo.") 
    money-=200

# eat jhalmuri 
elif money >= 10: 
    print("Jhalmuri khabo.")
    money-=10

else:
    print("Oh No!")

print(money)
````

## Nested Conditions

````python
# What is the problem in the following code?

money = 500 # 250, 50, 5 

# eat jhalmuri 
if money >= 10: 
    print("Jhalmuri khabo.")
    money-=10

# eat kacchi
elif money >=200: 
    print("Kacchi khabo.") 
    money-=200

# eat pizza
elif money >=300:
    print("Pizza khabo.") 
    money-=300

else:
    print("Oh No!")
````

````python
money = 500 # 250, 50, 5 

if money >= 10:
    if money >=200: 
        if money >=300:
            print("Pizza khabo.") 
            money-=300
        else:
            print("Kacchi khabo.") 
            money-=200
    else:
        print("Jhalmuri khabo.")
        money-=10
  
else:
    print("Oh No!")

print(money)
````

### Logical Operator instead of Nested Conditions

````python
money = 500 # 250, 50, 5 

if money >= 10 and money < 200: # 10 <= money <200
    print("Jhalmuri khabo.")
    money-=10

elif money >=200 and money < 300: # 200 <= money <300
    print("Kacchi khabo.") 
    money-=200

elif money >=300:
    print("Pizza khabo.") 
    money-=300
 
else:
    print("Oh No!")

print(money)
````

# print() function (usage of sep and end argument)

````python
var1 = "Hello"
var2 = "Students"

print(var1, var2)
print(var1, var2, sep = " ") # by default, the separator is a space
print(var1, var2, sep = "!")
````

````python
var1 = "Hello"
var2 = "Students"

print(var1)
print(var2)

print(var1, end = ", ")
print(var2)
````

````python
var1 = "Hello"
var2 = "Students"

print(var1)
print(var2)

# by default, Python always moves the cursor to the next line ("\n")
print(var1, end = "\n")
print(var2, end = "\n")
````

````python
var1 = "Hello"
var2 = "Students"

print(var1, end = "-") # prints the value and keeps the cursor in the current line
print(1) # prints 1 in the current line and moves the cursor to the next line
print(var2) # prints the value and moves the cursor to the next line
````

````python
var1 = "Hello"
var2 = "Students"

print(var1) # prints the value and moves the cursor to the next line
print() # prints a blank line and moves the cursor to the next line
print(var2) # prints the value and moves the cursor to the next line
````

# Coding Problem:
Suppose, as a product delivery service manager, you have to calculate the total cost combining the product cost and delivery charge. Write a Python program that takes two inputs: **product cost & location**; from the user and calculate the total cost. Based on the location, the delivery charge will vary. 

*   If the location is **"Mohakhali"**, then no delivery charge.
*   If the location is **"Banani"** or **"Gulshan"**, then the delivery charge will be **5%** of the product cost.
*   Otherwise, the delivery charge will be **10%** of the product cost. 

Finally, display the total cost. 

````python
prod_cost = float(input("Please Enter the Product Cost: ")) # type can be int as well
loc = input("Please Enter Location: ")
if loc == "Mohakhali":
    del_charge = 0
elif loc == "Banani" or loc == "Gulshan":
    del_charge = (prod_cost * 5)/100 # or, prod_cost * 0.05
else:
    del_charge = (prod_cost * 10)/100 # or, prod_cost * 0.1
total_cost = prod_cost + del_charge
print("Total Cost is:", total_cost)
````




````python
# Solution: 

res1 = True
res2 = True
res3 = False
res4 = False
res5 = True
````

````python
# Detailed Breakdown:
 
var1 = 20 < 17 # F

var2 = not False # T

var3 = True or not (20==20) # T or not T = T or F = T 
## shortcut: T or [anything] = T 

var4 = (5 > 7) and (1 < 2) # F and T = F 
## shortcut: F and [anything] = F

var5 = var1 or (2**2 != 9) and (23 % 5 == 5)
# F or T and F = F or (T and F) = F or F = F

# print(var1)
# print(var2)
# print(var3)
# print(var4)
# print(var5)


res1 = not var1 and var3 # not F and T = T and T = T
res2 = var2 or (10 in [100, 200, 500]) # T or F = T
res3 = var4 or (5 is "5") and var3 # F or F and T = F or F = F

print(res1)
print(res2)
print(res3)

var3 = not res2 # not T = F
var5 = res1 and res2 # T and T = T

# print(var3)
# print(var5)

res4 = (res1 and var3) or not(var5 and res1) 
# (T and F) or not(T and T) = F or not T = F or F = F

res5 = (res4 or not res1) and (not res1 or res2) or (not res3)
# (F or not T) and (not T or T) or (not F) 
# = (F or F) and (F or T) or T 
# = F and T or T = F or T = T

print(res4)
print(res5)
````

## Slicing


*   Denotes accessing parts of a sequence
*   [start = 0 : stop *(exclusive)* = len : step = 1]

````python
str1 = "First Quiz"

print(str1[::]) # First Quiz
print(str1[:])  # First Quiz

print(str1[0:5]) # First
print(str1[0:5:3]) # Fs
print(str1[0::3]) # FsQz 

print(str1[6:]) # Quiz
print(str1[-4:]) # Quiz

# reverse
print(str1[::-1]) # ziuQ tsriF
````

## Problem: 
Take a sentence from the user as input. Your task will be to modify the sentence in such a way that all the digits are removed and all the lowercase letters are converted into uppercase. Finally, display the modified sentence. <br>
[You can not use any built-in function for strings except ord() and chr() in this problem]

**Sample Input:** CSE110 Quiz 1 will be easy <br>
**Sample Output:** CSE QUIZ&nbsp;&nbsp;WILL BE EASY

````python
str1 = input()
str2 = ""
for char in str1:
    if 97 <= ord(char) <= 122:
        str2 += chr(ord(char)-32)
    elif 48 <= ord(char) <= 57:
        continue
    else:
        str2 += char
print(str2)
````

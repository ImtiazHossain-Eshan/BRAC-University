# Variables

*   General Definition: Something that changes or can change. In other words, something which is not constant. 
*   In Python, a variable can be compared with a container.
*   Variable is a reserved memory location to store values. 
*   No need to declare first like other languages. 
*   Declaration Format: *variable_name = value*
*   Assigned from right to left.


```python
age = 18
print(age)

age = 19.5
print(age)

age = "Teen"
print(age)

del age # delete a variable
print(age)
```

## Variable Naming Conventions: 
For better understanding & readability, it is recommended to follow a naming convention.


1.   It is better to begin with lowercase letters. (age)
2.   A variable can not have whitespace. For more than one word variables, use seperator '_' between two words or capitalize the first letter of the later words. (student_name or studentName)
3.   Never start with digits. (student1 is okay but 1student is wrong)
4.   A variable can not have special signs (@, $, %, +, -, ?, ! etc) 

## Some Instructions:
1.   Try to make it as meaningful as possible. 
2.   Neither too short (a, b, x, y) nor too long (my_favourite_color_name)
3.   Do not use 'I' (or 'l') or 'O' (or 'o') because it looks almost similar to 1 and 0 respectively

*   Mapping Type (unordered collection in key:value pair): dict

# Operators 

## Arithmatic Operators


1.   Addition **+**
2.   Subtraction **-**
3.   Multiplication (Product) * 
4.   Division **/** (result will always be in float)
5.   Modulus **%**
6.   Floor division **//** (need to know difference between ceiling and floor)
7.   Exponentiation (Power) **

```python
print(6 + 2) # 8
print(6 - 2) # 4
print(6 * 2) # 12
print(6 / 2) # 3.0
print(6 % 2) # 0
print(6 // 2) # 3 
print(6 ** 2) # 36
```

```python
print(10 / 2) 
print(10.0 / 2)
```

To understand floor division (//), you need to understand the concept of floor (and ceiling) in mathematics. 


*   ceiling(2.4) = 3 
*   floor(2.9) = 2
*   ceiling(-2.4) = -2 
*   floor(-2.9) = -3

```python
print(10 // 2) 
print(10.0 // 2) 

print(11 // 4) 
print(11.0 // 4)
```

```python
print(11 % 2) 
print(11.5 % 2)
```

```python
print(1 / 2) # 0.5
print(1 // 2) # 0
print(1 % 2) # 1  [x % y = x ; if x is less than y]
```

```python
print(1.0 / 2) # 0.5
print(1.0 // 2) # 0.0
print(1.0 % 2) # 1.0
```

```python
# [x % y = x - y * (x // y) ; if x and/or y is negative]
print(-5 % 2) # 1 
print(5 % -2) # -1
print(-5 % -2) # -1
```

```python
print(-7 // 2) # -4
print(-7 % 2) # 1
print(7 // -2) # -4
print(7 % -2) # -1
print(-7 // -2) # 3
print(-7 % -2) # -1
```

## Assignment Operators

Assignment Operators: =, +=, -=, * =, /=, %=, ** =, //= 

*General Format:*

*variable = variable (some operation) number or another variable*

==> *variable (some operation)= number or another variable*

```python
x = 2
x = x + 3
print(x) # 5

x = 2
x += 3
print(x) # 5 (same as before)
```

```python
x = 2
x = x * 3
print(x) # 6

x = 2
x *= 3
print(x) # 6
```

## Identity Operators

Identity Operator: **is, is not**

Checks if two objects (type and value both) are same or not (result is boolean)

```python
print(5 is 5)
print(5 is 5.0)
print(5 is 6) 
print(5 is "5") 
```

```python
print(5 is not 5) 
print(5 is not 6) 
print(5 is not "5")
```

```python
a = 5
print(a is 5) 
print(a is not 5) 
```

### An Exception 
*   Do not use Identity Operators to compare between two values. <br>
*   It may not give us desired output in some cases like below. <br>
*   Instead we can use Comparison or Relational Operators (which we will study in the later part).





```python
print(0.5 is 0.5)
```

```python
x = 0.5
y = 0.5
print(x is y) 
```

```python
x = 0.5
y = x
print(x is y)
```

## Membership Operators

Membership Operators: **in, not in**

Checks if a value is present in a sequence or not (result is boolean)

```python
print("y" in "Python")
print("b" in "Python")
```

```python
print("y" not in "Python") 
print("b" not in "Python") 
```

```python
print("hi" in ["hi","hello"])
print("hola" in ["hi","hello"])
```

```python
print("hi" not in ["hi","hello"]) 
print("hola" not in ["hi","hello"])
```

# **Sorting**

Sorting means rearranging values in a particular order.
<br>
<br>
Two types: 
1.   Ascending Order / Increasing Order / Smallest to Largest / "Sort" (by default)
2.   Descending Order / Decreasing Order / Non-increasing Order / Largest to Smallest 


## **Built-in Sort**

```python
# sort() function (in-place sort)
list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print(list1)
list1.sort()
print(list1) 

print()

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print(list1)
list1.sort(reverse = True)
print(list1) 
```

```python
# sorted() function (out of place sort / creates a new list)

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
list2 = sorted(list1)
print(list1)
print(list2)

print()

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
list2 = sorted(list1, reverse = True)
print(list1)
print(list2)
```

```python
# sorted() in String 
str1 = "Hello World"
str2 = sorted(str1)
str3 = sorted(str1, reverse = True)
print(str1)
print(str2)
print(str3)
```

```python
# sorted() in Dictionary  
dict1 = {'Tobey': 46, 'Andrew': 38, 'Tom': 26}
dict2 = sorted(dict1) # generates the sorted key list 
print(dict2) # type list

print()

for i in dict1:
    print(i, dict1[i])

print()

for i in sorted(dict1): # dict2
    print(i, dict1[i])
```

## **Selection Sort**
*   assume, the current index has the smallest (for ascending order sort) / largest (for descending order sort) value
*   searches from (current index + 1) up to (length - 1) for the index with the smallest / largest value
*   swaps the value of the current index with the value of the index with the smallest / largest value
*   starting from the left-most index, the values from each index get sorted one by one after each iteration 

```python
def selectionSortAscending(lst):
    for curr_i in range(len(lst)):
        m_idx = curr_i
        for j in range(curr_i+1, len(lst)):
            if lst[j] < lst[m_idx]:
                m_idx = j
        # swapping
        temp = lst[curr_i] 
        lst[curr_i] = lst[m_idx]
        lst[m_idx] = temp
        # lst[curr_i] , lst[m_idx] = lst[m_idx], lst[curr_i]
    return lst

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print("Before Sorting:",list1)
print("After Sorting:",selectionSortAscending(list1))
```

```python
def selectionSortDescending(lst):
    for curr_i in range(len(lst)):
        m_idx = curr_i
        for j in range(curr_i+1, len(lst)):
            if lst[j] > lst[m_idx]:
                m_idx = j
        # swapping
        temp = lst[curr_i] 
        lst[curr_i] = lst[m_idx]
        lst[m_idx] = temp
        # lst[curr_i] , lst[m_idx] = lst[m_idx], lst[curr_i]
    return lst

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print("Before Sorting:",list1)
print("After Sorting:",selectionSortDescending(list1))
```

## **Bubble Sort**
*   main idea is to check and swap between consecutive two index values
*   starting from the right-most index, the values from each index get sorted one by one after each iteration

```python
def bubbleSortAscending(lst):
    for i in range(len(lst)-1): # range(len(lst))
        for j in range(len(lst)-1-i): # range(len(lst)-1)
            if lst[j+1] < lst[j]:
                # swapping
                temp = lst[j] 
                lst[j] = lst[j+1]
                lst[j+1] = temp
                # lst[j] , lst[j+1] = lst[j+1] , lst[j]
    return lst

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print("Before Sorting:",list1)
print("After Sorting:",bubbleSortAscending(list1))
```

```python
def bubbleSortDescending(lst):
    for i in range(len(lst)-1): # range(len(lst))
        for j in range(len(lst)-1-i): # range(len(lst)-1)
            if lst[j+1] > lst[j]:
                # swapping
                temp = lst[j] 
                lst[j] = lst[j+1]
                lst[j+1] = temp
                # lst[j] , lst[j+1] = lst[j+1] , lst[j]
    return lst

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print("Before Sorting:",list1)
print("After Sorting:",bubbleSortDescending(list1))
```

# **Searching**

## **Linear Search / Sequential Search**

```python
def linearSearch(lst, val):
    for i in range(len(lst)):
        if lst[i] == val:
            return "Value found at index " + str(i)
    return "Value not found"

list1 = [10, 50, 27, -5, 3, 12, -2, 31]
print(linearSearch(list1, 12))
print(linearSearch(list1, 25))
```

## **Binary Search**

### **https://www.cs.usfca.edu/~galles/visualization/Search.html**
(Animation for Binary Search, *Do not use this site for understanding Linear Search*)

### Pre-requisite: **The values MUST be sorted**

```python
def binarySearch(lst, val):
    st = 0
    end = len(lst) - 1
    while st<=end:
        mid = (st+end)//2
        # print(st, end, mid)
        if lst[mid] == val:
            return "Value found at index " + str(mid)
        elif lst[mid] < val:
            st = mid + 1
        else:
            end = mid - 1
    return "Value not found"

list1 = [-5, -2, 3, 10, 12, 27, 31, 50]
print(binarySearch(list1, 27))
print(binarySearch(list1, 25))
```
